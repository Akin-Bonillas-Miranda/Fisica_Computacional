*`Resuelva la ecuación diferencial del`* [*`oscilador de Van der Pol`*](https://en.wikipedia.org/wiki/Van_der_Pol_oscillator)

\begin{equation*}
\frac{d^2 x}{d t^2} - \mu (1 - x^2) \frac{dx}{dt } + x = 0
\end{equation*}

*`donde`* $x$ *`es la posición y`* $\mu$ `es un parámetro de la parte no lineal. `

*`Resuelva el caso para`* $\mu = 0, 1, 2, 3, 4$ *`y grafique las soluciones para un tiempo de integración de`* $t = [0, 50]$


\begin{equation}
  \frac{dx}{dt} = v \rightarrow \frac{ dv }{ dt } = \mu( 1 - x^2 )v -x
\end{equation}

```
#==================== OSCILADOR DE VAN DER POL PARA MU = 0 ====================#

from scipy.integrate import solve_ivp, odeint

# definimos la función F de lado derecho de la ecuación.
def OslVDP( y, t, mu ):
        x, v = y
        dydt = [ v, mu*(1 - x**2)*v - x ]
        return dydt

t = np.linspace( 0, 50, 501 ) # Definimos el tiempo de integración

mu = 0 # Definimos el coeficiente mu

t_0 = 0.0        # Definimos las condiciones iniciales 
y_0 = [1.0, 0.0]

Solve = odeint( OslVDP, y_0, t, args=(mu,) ) 

# La solución regresa ambas x(t) y v(t)
y1 = Solve [:,0]
y2 = Solve [:,1]

plt.figure( figsize = (18,6) )
plt.plot( t, y1, "b", label = "x(t)" )
plt.plot( t, y2, "g", label = "v(t)" )
plt.legend( loc = "best" )
plt.title( "Solución del Oscilador de Van der Pol, mu=0" )
plt.xlabel( "t" )

plt.grid()
plt.show()
```

\begin{equation*}
\frac{d^2 x}{dt^2} + k (\frac{dx}{dt})^2 -g = 0, \ x(0)=0, \ \frac{dx}{dt}(0). = 0, \ k>0.
\end{equation*}

*`Grafique las soluciones como funciones del tiempo.`*

*`Para facilitar los cálculos a realizar podemos reescribir la ecuación diferencial de la siguiente forma:`*

\begin{equation}
  \frac{dx}{dt} = v = f_{1}(t) \\

  \frac{dv}{dt} = g - kv^{2} = f_{2}(t)
\end{equation}

*`Estos resultados podemos expresarlos cómo:`*

\begin{equation}
  \frac{d}{dt}\vec{y} = \vec{F}( t, \vec{y}) = ( f_{1}(t), f_{2}(t) )
\end{equation}

*`Donde tenemos que:`*

\begin{equation*}
  \vec{y} = \begin{bmatrix}
    x \\
    v \\
  \end{bmatrix}
\end{equation*}

```
#===================== REESCRIBIMOS LA ECUACIÓN DIFERENCIAL ====================

def EDO1( y, t, k, g ):
        x, v = y
        dydt = [v, g - k*v**2 ]
        return dydt

t = np.linspace(0, 4, 51)
t0 = 0
x0 = 0
v0 = 0

k = 1
g = 9.81
```

```
#========================= APLICAMOS EL MÉTODO DE EULER ========================

#------------------------ Resolvemos la primera ecuación -----------------------

f = lambda v,t: -k*v**2 + g 
v = np.zeros( len( t ) )
v[ 0 ] = v0
    
for n in range( 0, len( t ) - 1 ):
  v[ n+1 ] = v[ n ] + f (v[ n ], t[ n ] )*( t[ n+1 ] - t[ n ] )

#------------------------ Resolvemos la segunda ecuación -----------------------

x = np.zeros( len(t) )
x[ 0 ] = x0
    
for n in range(0,len(t)-1):
  x[ n+1 ] = x[ n ] + v[ n ]*( t[ n+1 ] - t[ n ] )

plt.figure( figsize = ( 10, 5 ) )
plt.plot( t, v, "b.-", t, x, "g.-" )
plt.legend( [ "$v(t)$", "$x(t)$" ] )
plt.title( "Solución de la ecuación diferencial con Euler" )
plt.axis( [ 0, 2, 0, 9 ] )
plt.grid()
plt.xlabel( "t" )
plt.show()
```

```
#============================ MÉTODO DE RUNGE-KUTTA ============================

def MRK( f, x_0, y_0, h ):
  
    k_0 = f(x_0, y_0)
    k_1 = f(x_0 + h/2, y_0 + h/2 * k_0)
    k_2 = f(x_0 + h/2, y_0 + h/2 * k_1)
    k_3 = f(x_0 + h, y_0 + h * k_2)

    k = 1/6 * (k_0 + 2.0*k_1 + 2.0*k_2 + k_3)

    x_1 = x_0 + h
    y_1 = y_0 + h * k

    return x_1, y_1
```

# ***`Ejercicio 3.2:`***

\begin{equation*}
\frac{d^2 y}{d x^2} + \frac{2}{x} \frac{dy}{dx} + y^5 =0, \ y(0) = 1, \ \frac{dy}{dx}(0)=0. 
\end{equation*}

*`Grafique las soluciones como funciones del tiempo.`*

*`Para facilitar los cálculos a realizar podemos reescribir la ecuación diferencial de la siguiente forma:`*

\begin{equation}
  \frac{dx}{dt} = v = f_{1}(t) \\

  \frac{dv}{dt} = - \frac{2}{x}v - y^{5} = f_{2}(t)
\end{equation}

*`Estos resultados podemos expresarlos cómo:`*

\begin{equation}
  \frac{d}{dt}\vec{a} = \vec{F}( t, \vec{a}) = ( f_{1}(t), f_{2}(t) )

*`Donde tenemos que:`*

\begin{equation*}
  \vec{a} = \begin{bmatrix}
    y \\
    v \\
  \end{bmatrix}
\end{equation*}

*`Dado que las ecuaciones de primer orden dependen de más de 2 variables no es posible aplicar ni el método de Euler, ni el método de Runge-Kutta`*

\end{equation*}

*`Grafique las soluciones como funciones del tiempo.`*

*`Para facilitar los cálculos a realizar podemos reescribir la ecuación diferencial de la siguiente forma:`*

\begin{equation}
  \frac{dx}{dt} = v = f_{1}(t) \\

  \frac{dv}{dt} = a = f_{2}(t)\\

\end{equation}

\begin{equation}
  \frac{d}{dt}\vec{w} = \vec{F}( t, \vec{w}) = ( f_{1}(t), f_{2}(t), f_{3}(t) )
\end{equation}

*`Donde tenemos que:`*

\begin{equation*}
  \vec{w} = \begin{bmatrix}
    y \\
    v \\
    a \\
  \end{bmatrix}
\end{equation*}

*`Dado que las ecuaciones de primer orden dependen de más de 2 variables no es posible aplicar ni el método de Euler, ni el método de Runge-Kutta`*

```
#================= APLICACAMOS EL MÉTODO scipy.integrate.odeint ================

def EDO3( b, x ):
  y, v, a = b
  dbdx = [ v, a, ( x - 1 )**2 + y**2 + v - 2]
  return dbdx

x0 = 0.0
y0 = 1.0
v0 = 0.0
a0 = 2.0
x = np.linspace( 0.0, 2.0, 51)

x_0 = x0             # Definimos las condiciones iniciales 
a_0 = [ y0, v0, a0 ]

abserr = 1.0e-8 # Definimos los errores de tolerancia para la solución
relerr = 1.0e-6

Solve = odeint( EDO3, a_0, x, atol = abserr, rtol = relerr )

y1 = Solve[:,0] # La solución regresa ambas x(t) y v(t)
y2 = Solve[:,1]
y3 = Solve[:,2]

plt.figure( figsize = ( 10, 5 ) )
plt.plot( x, y1, "b.-", label = "$x(t)$" )
plt.plot( x, y2, "g.-", label = "$v(t)$" )
plt.plot( x, y3, "r.-", label = "$a(t)$" )
plt.legend( loc = "best" )

plt.title( "Solución de la ecuación diferencial con odeint" )
plt.xlabel( "t" )
plt.grid()
plt.show()
```
